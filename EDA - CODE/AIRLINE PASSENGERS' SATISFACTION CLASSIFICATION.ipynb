##### Plot the data before filling null

```python
sns.set_palette("husl")
sns.displot(data=data, x="Arrival Delay", kind="kde")
```

##### FILLING THE NULLS

```python
data["Arrival Delay"] = data["Arrival Delay"].fillna(data["Arrival Delay"].mean())
```

```python
sns.set_palette("husl")
sns.displot(data=data, x="Arrival Delay", kind="kde")
```

### IDENTIFYING AND HANDLING THE OUTLIERS

##### PLOTTING BOX PLOT & SCATTER PLOT (FLIGHT DISTANCE vs SATISFACTION)

```python
sns.set_palette("Set3")
bp = sns.boxplot(data=data, x="Flight Distance")
plt.show()
sp = sns.scatterplot(data=data, x="Flight Distance", y=data["Satisfaction"])
plt.show()
```

##### IDENTIFYING THE INTERQUARTILE RANGE (IQR) & DROPPING THE OUTLIERS

```python
Q1_fd = data['Flight Distance'].quantile(0.25)
Q3_fd = data['Flight Distance'].quantile(0.75)
IQR_fd = Q3_fd - Q1_fd
UB_fd = Q3_fd + 1.8*(IQR_fd)
data.drop(data[data['Flight Distance']>UB_fd].index, axis=0, inplace=True)
```

```python
bp = sns.boxplot(data = data, x = "Flight Distance")
plt.show()
sp = sns.scatterplot(data = data, x = "Flight Distance", y=data["Satisfaction"])
plt.show()
```

##### PLOTTING BOX PLOT & SCATTER PLOT (DEPARTURE DELAY vs SATISFACTION)

```python
bp = sns.boxplot(data=data, x="Departure Delay")
plt.show()
bp = sns.scatterplot(data=data, x="Departure Delay", y=data["Satisfaction"])
plt.show()
```

##### REMOVING OUTLIER

```python
#removing outlier
data.drop(data[data['Departure Delay']>550].index, axis=0, inplace=True)
```

```python
bp = sns.boxplot(data=data, x="Departure Delay")
plt.show()
bp = sns.scatterplot(data=data, x="Departure Delay", y=data["Satisfaction"])
plt.show()
```

##### PLOTTING BOX PLOT & SCATTER PLOT (ARRIVAL DELAY vs SATISFACTION)

```python
bp = sns.boxplot(data=data, x="Arrival Delay")
plt.show()
bp = sns.boxplot(data=data, x="Arrival Delay", y=data["Satisfaction"])
plt.show()
```
